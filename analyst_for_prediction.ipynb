```python
import pandas as pd #mengimport fungsi pandas (library)

df = pd.read_csv("/content/drive/My Drive/dataset/heart.csv") #membaca dataset yang tersimpan di gdrive
df.head(4) #membaca 5 baris dari dataset
```

```python
a2 = df[(df['oldpeak'])==0.1]
a2
```

```python
a1 = a1 = df[(df['age']>40)&(df['age']<50)] #menampilkan data yang berumur >40 tahun dan >50 tahun 
a1
```

```python
x = df.iloc[:,0:13] #penggunaan fungsi iloc untuk memisahkan data x dan y sesuai kolom yang digunakan pada dataset
y = df.iloc[:, 13] #':' mewakili seluruh kolom, '13' mewakili kolom ke 13
y
```

```python
from sklearn.model_selection import train_test_split #sklearn untuk membagi data training dan testing (mengacak posisi)
x_train, x_test, y_train, y_test = train_test_split (x,y, test_size=0.1) #x adalah data dan y adalah target dan membagi ke 4 parameter x,y train/test
```

```python
from sklearn.svm import SVC #model svm untuk klasifikasi data

clf = SVC() #membuat data SVC
clf = clf.fit(x_train, y_train) #training data dan data di simpan di clf
y_pred = clf.predict (x_test) #data diprediksi dan disimpan di y_pred | Model machine learning seseorang memiliki penyakit jantung berdasarkan 14 parameter
```

```python
from sklearn.metrics import accuracy_score #melakukan evaluasi dengan library metrics dengan akurasi skor
print (accuracy_score(y_pred, y_test)) #hasil akurasi 77%
```
